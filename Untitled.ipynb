```python
import matplotlib as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('SHARADAR_SF3A_e25b0a5f97511022da11f3ac8a1b2034.csv')
df=df[(df.shrholders!=0)&(df.shrvalue!=0)&(df.shrunits!=0)]
df.head()
```

```python
#callput ratio
cpratio=[]
ratio=0
for index, row in df.iterrows():
    if(row['cllunits']==0):
        ratio=0
    else:
        ratio=row['putunits']/row['cllunits']
    cpratio.append(ratio)
```

```python
#debt-equity ratio
deratio=[]
ratio=0
for index, row in df.iterrows():
    ratio=row['dbtvalue']/row['shrvalue']
    deratio.append(ratio)
```

```python
#percent preferred
preferred=[]
per=0
for index, row in df.iterrows():
    per=row['prfunits']/(row['shrunits']+row['prfunits'])
    preferred.append(per)
```

```python
#percent warrants
warrants=[]
per=0
for index, row in df.iterrows():
    per=row['wntunits']/(row['shrunits']+row['prfunits']+row['wntunits'])
    warrants.append(per)
```

```python
df2=df.assign(callPutRatio=cpratio).assign(debtEquityRatio=deratio).assign(percentPreferred=preferred).assign(percentWarrants=warrants)
df2.head()
```

```python
df2.to_csv("sharadar_new.csv")
```
